# Hierarchial Clustering (ward's method)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
# Load Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# Step 1: Perform hierarchical clustering (using Wardâ€™s method)
Z = linkage(data, method='ward')
```

```python
# Step 2: Plot dendrogram
plt.figure(figsize=(10, 6))
dendrogram(Z, truncate_mode='lastp', p=30, leaf_rotation=45., leaf_font_size=10., show_contracted=True)
plt.title('Hierarchical Clustering Dendrogram (Iris)')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.show()
```

```python
# Step 3: Cut dendrogram into 3 clusters
clusters = fcluster(Z, 3, criterion='maxclust')
```

```python
# Step 4: Visualize clusters (using only first two features)
sns.scatterplot(x=data.iloc[:,0], y=data.iloc[:,1], hue=clusters, palette='Set1')
plt.title('Hierarchical Clustering (Iris)')
plt.show()
```

Using hierarchical clustering on the Iris dataset, we identified 3 clusters. One cluster (red) is clearly separated, while the other two overlap partially. This matches the natural distribution of the Iris dataset, where Setosa is easily distinguishable, but Versicolor and Virginica are more similar and harder to separate. Thus, hierarchical clustering successfully uncovers the underlying grouping structure of the dataset
